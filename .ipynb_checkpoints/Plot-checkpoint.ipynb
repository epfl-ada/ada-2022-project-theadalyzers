```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(1,2, figsize=(18,8), sharey=True)

vis_f=pd.read_csv('vis_f.csv', index_col=0)
vis_u=pd.read_csv('vis_u.csv', index_col=0)
vis=pd.concat([vis_f,vis_u], axis=0)

sns.barplot(y='#_clicks / #_hyperlinks', x=vis.index, data=vis, palette=sns.color_palette("Set2"), hue='is_successful', edgecolor='k', ax=ax[0])
#ax[0].set_yscale("log")

ax[0].set_xlabel('Positioning', fontsize=18)
ax[0].set_ylabel('#_clicks / #_hyperlinks', fontsize=20)
ax[0].legend_.set_title('')

ax[0].xaxis.set_tick_params(labelsize = 14)
ax[0].yaxis.set_tick_params(labelsize = 18)

labels = ['Unsuccessful', 'Successful']
for t, l in zip(ax[0].legend_.texts, labels):
    t.set_text(l)



vis_img=pd.read_csv('vis_img.csv')
vis_img.set_index('in_image', inplace=True)

sns.barplot(y='#_clicks / #_links_img', x=vis_img.index, data=vis_img, palette=sns.color_palette("Set2"), hue='is_successful', edgecolor='k', ax=ax[1])
#ax[1].set_yscale("log")

ax[1].set_xlabel('In Image', fontsize=18)
ax[1].set_ylabel('', fontsize=18)
ax[1].legend_.set_title('')
ax[1].xaxis.set_tick_params(labelsize = 14)
labels = ['Unsuccessful', 'Successful']
for t, l in zip(ax[1].legend_.texts, labels):
    t.set_text(l)

fig.suptitle('Distribution of clickability preference', fontsize=22)
plt.savefig('Distribution_pos_img.png', bbox_inches='tight')
```
