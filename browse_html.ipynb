Code for extracting the clickthrough rate of hyperlinks

```python
# Creation of the list of paths
# For each path, creates a list with the clicked links

clicked_links = []
for x in paths_finished_df['path']:
    path = (x.split(';')) # list of clicked links in each path
    clicked_links.append(path)

clicked_links[:10]
```

```python
# Creation of the dataframe containing the hyperlinks and their click-through rates

links_name=[] # list of all articles' names
links_freq=[]

for article in art_df['article']:
    links_name.append(article)
    count=0
    for i in range(len(clicked_links)):
        if article in clicked_links[i]:
            count+=1 
    links_freq.append(count)

links_name = pd.Series(links_name)
links_freq = pd.Series(links_freq)
links_freq_df = pd.concat([links_name, links_freq], axis=1, 
                          names=['link', 'clickthrough_rate'])
links_freq_df.columns = ['link', 'clickthrough_rate']
links_freq_df.head(10)
```

```python
# Extract the positioning of the hyperlinks in the article text

from bs4 import BeautifulSoup

URL = '/home/Project/wpcd/wp/a/A_cappella.htm'

f=open(URL, 'r')
soup = BeautifulSoup(f, 'lxml')

paragraphs = soup.find_all('p') # each paragraph is located by the tag 'p'
print(f'The total number of paragraphs is {len(paragraphs)}.')
```

```python
# HTML tag for hyperlinks 
# This is the dataframe of the hyperlinks contained in an article, with the 
# number of paragraph they show up in and its relative position in the article

links=[] # list of links' titles and paragraphs
i=0

for p in paragraphs:
    i+=1
    all_links = p.find_all('a') # each hyperlink is located by the tag 'a'
    for link in all_links:
        if 'href' in link.attrs: # href is the attribute indicating the link's destination
            link_title = link.text
            link_paragraph = i
            links.append([link_title, link_paragraph])

link_positioning_df = pd.DataFrame(links, columns=['link', '#_paragraph'])
P = len(paragraphs)
link_positioning_df['positioning'] = round(link_positioning_df['#_paragraph']/P, 2)
link_positioning_df.head()
```

```python
# Find hyperlinks in images' caption

URL = '/home/Project/wpcd/wp/a/Aachen.htm'
f2 = open(URL, 'r')
soup2 = BeautifulSoup(f2, 'lxml')
```

In an HTML file the images can be found either under **tables** or **div** tags

```python
tables = soup2.find_all('table')
print(f'The total number of tables is {len(tables)}.')
```

```python
img = soup2.find_all('img')
print(f'The total number of images is {len(img)}.')
```

```python
# Images under table tags
# Dataframe of hyperlinks in tables' captions

links = []

for table in tables: # ???
    if (table.img):
        all_links = table.find_all('a')
        for link in all_links:
            if 'class' not in link.attrs:
                link_title = link.text
                links.append(link_title)
        
links_in_table_df = pd.DataFrame(links, columns=['link_in_table'])
links_in_table_df
```

```python
# images under div tags
# dataframe of hyperlinks in tables' captions

div = soup2.find_all('div')

links=[]

for d in div:
    if (d.find('a', class_="internal")):
        if (d.find('img', class_='thumbimage')):
            caption = d.find('div', class_="thumbcaption")
            if (caption.find('a')):
                all_links = caption.find_all('a')
                for link in all_links:
                    if 'class' not in link.attrs:
                        link_title = link.text
                        links.append(link_title)
                            
links_in_image_df = pd.DataFrame(links, columns=['link_in_image']).drop_duplicates() 
links_in_image_df    
```
